# **Contexte**

Olist, a sales solution on online marketplaces, wants to provide their e-commerce teams with customer segmentation that they can use on a daily basis for their communication campaigns.

The goal is to understand the different types of users through their behavior and their personal data.
We will need to provide the marketing team with an actionable description of our segmentation and its underlying logic for optimal use, as well as a maintenance contract proposal based on an analysis of the stability of the segments over time.

For this mission, Olist has provided us with an anonymized database with information on order history, products purchased, satisfaction reviews, and customer location since January 2017.
Our mission is to help the Olist teams understand the different types of users. 

We will therefore use unsupervised methods to group together clients of similar profiles. These categories can be used by the marketing team to communicate better.
Finally, the client, Olist, specified their request as follows:
The proposed segmentation must be actionable and easy to use for the marketing team.

We will assess how often the segmentation needs to be updated, in order to be able to perform a maintenance contract quote.

In this first part we will perform file cleaning and a short exploratory analysis. In the second part, we will carry out the various models in order to realise a customer segmentation using unsupervised methods.

they are Web pages accessed, or answers to multiple-choice questions from the registration questionnaire. Most of the customers have only placed one order. It will therefore be 
complicated to establish a classification of their preferred product 
category. We are therefore going to create a variable per product 
category to dispatch the number of products purchased on the orders 
placed. Next, we will calculate the ratio of the number of products per 
category to the total number of products purchased by the customer.

```python
categories_customers = data.groupby(["customer_unique_id", "product_category"])\
                        .agg({"order_item_id": "count"}).unstack()
categories_customers.columns = categories_customers.columns.droplevel(0)
categories_customers.fillna(0, inplace=True)
categories_customers["total_items"] = categories_customers.sum(axis=1)

# ratio of total items
for col in categories_customers.columns:
    if (col != "total_items"):
        categories_customers[col] = (
            categories_customers[col]/categories_customers["total_items"])

categories_customers.reset_index(inplace=True)
categories_customers.head(5)
```

```python
# Drop product_category_name column
data.drop("product_category_name", axis=1, inplace=True)
```

**Creation of the customer dataset**

```python
# Group by customers with sum or mean
data = data.groupby("customer_unique_id")\
            .agg({"order_id": "nunique",
                  "price": "sum",
                  "freight_value": "sum",
                  "nb_payment_sequential": "mean", 
                  "sum_payment_installments": "mean", 
                  "review_score": "mean",
                  "delivery_days": "mean",
                  "order_purchase_timestamp": "mean",
                  "sale_month": lambda x:x.value_counts().index[0]})

# Rename columns
data = data.rename(columns={"order_id": "nb_orders",
                            "price": "total_spend",
                            "freight_value": "total_freight",
                            "nb_payment_sequential": "mean_payment_sequential",
                            "sum_payment_installments": 
                            "mean_payment_installments",
                            "review_score": "mean_review_score",
                            "delivery_delta_days": "mean_delivery_days",
                            "order_purchase_timestamp": "order_time",
                            "sale_month": "favorite_sale_month"})
# Merge with other calculated features
data = pd.merge(data, categories_customers,
                how="left",
                on="customer_unique_id")


data.head(6)
```

In order to use an encodable location variable without negative impact 
on our models, we will calculate the Haversine distance between the 
client's state (average of the state's latitudes and longitudes) and 
the seat of Olist:

The Haversine (or great circle) distance is the angular distance 
between two points on the surface of a sphere. The first coordinate of 
each point is assumed to be the latitude, the second is the longitude, 
given in radians. The dimension of the data must be 2.

```python
state_location = data_geolocation.groupby("geolocation_state").mean()
state_location = pd.merge(states_table[["Flag and name"]].reset_index(), 
                          state_location, 
                          right_on="geolocation_state", 
                          left_on="Code", 
                          how="left")
state_location = state_location[["Flag and name", "geolocation_lat",
                                 "geolocation_lng"]]

# Olist location
olist_lat = -25.43045
olist_lon = -49.29207
```

```python
def haversine_distance(lat1, lng1, lat2, lng2, degrees=True):
    r = 3956 # rayon de la Terre en miles
    
    if degrees:
        lat1, lng1, lat2, lng2 = map(radians, [lat1, lng1, lat2, lng2])
    
    # Formule Haversine
    dlng = lng2 - lng1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlng/2)**2
    d = 2 * r * asin(sqrt(a))  

    return d
```

```python
# Haversine distance between centre state and Olist office
from math import radians, cos, sin, asin, sqrt
state_location['harvesine_distance'] = [haversine_distance(olist_lat, olist_lon,
                                                           x, y) 
                              for x, y in zip(state_location.geolocation_lat, 
                                              state_location.geolocation_lng)]
```

```python
data = pd.merge(data, state_location[["Flag and name", "harvesine_distance"]],
         how="left",
         left_on="customer_state",
         right_on="Flag and name")
data.set_index("customer_unique_id", inplace=True)
data.head(5)
```
